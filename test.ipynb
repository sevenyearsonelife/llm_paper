```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

def find_top_similar(embedding, embedding_list, top_n=5):
    """
    找到与给定embedding余弦相似度最高的top_n个向量

    参数：
    embedding (list or numpy array): 目标向量
    embedding_list (list of lists or numpy array): 向量列表
    top_n (int): 返回的最相似向量的数量

    返回：
    numpy array: 与目标向量余弦相似度最高的top_n个向量
    """
    # 将输入转换为numpy数组
    embedding = np.array(embedding)
    embedding_list = np.array(embedding_list)

    # 计算余弦相似度
    similarities = cosine_similarity([embedding], embedding_list)[0]

    # 找到相似度最高的top_n个向量的索引
    top_n_indices = np.argsort(similarities)[-top_n:][::-1]

    # 获取相似度最高的top_n个向量
    top_n_embeddings = embedding_list[top_n_indices]

    return top_n_embeddings


def find_top_similar_paragraphs(query_index, top_n):
    # 示例变量
    embedding = generate_embeddings(paragraphs_query[query_index])
    
    # 调用函数
    top_similar_embeddings = find_top_similar(embedding, embedding_list, top_n=top_n)

    # 输出结果
    print("问题:", paragraphs_query[query_index])
    print("答案:", query_answer_dict[paragraphs_query[query_index]])
    print("------------")
    for emb in top_similar_embeddings:
        text = embedding_to_text[str(list(emb))]
        print(text)

# 示例调用
find_top_similar_paragraphs(4, top_n=5)


```
